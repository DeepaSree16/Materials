Check if a number is a circular prime

A circular prime is a prime number with the property that the number generated at each intermediate step when cyclically permuting its digits will be prime. For example, 1193 is a circular prime, since 1931, 9311 and 3119 all are also prime.

```python
def is_prime(n: int) -> bool:
    if n == 2 or n == 3:
        return True
    if n % 2 == 0:
        return False
    factor = 3
    while factor * factor <= n:
        if n % factor == 0:
            return False
        factor += 2
    return True
```

```python
def circularize_num(n: int) -> int:
    s = str(n)
    return int(s[1:] + s[0])
```

```python
def is_circular_prime(n: int) -> bool:
    start = n
    count = 0
    while count <= len(str(n)):
        n = circularize_num(n)
        if not is_prime(n):
            return False
        count += 1
    return True
```
