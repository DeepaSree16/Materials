We start by noting that we have done all the ground work by decoupling the pattern generation from formatting





```python
STAR, SPACE, CRLF = '*', ' ', '\n'

START, REPEAT, END = STAR, SPACE + STAR, ''

WIDTH = 60

def line(line_num: int) -> str:
    return start_line(line_num) + repeat_middle(line_num) + end_line(line_num)

def start_line(line_num: int) -> str:
    return START

def repeat_middle(line_num: int) -> str:
    return line_num * REPEAT

def end_line(line_num: int) -> str:
    return END

def make_pyramid(size: int) -> [str]:
    pyramid = []
    for line_num in range(size):
        pyramid.append(line(line_num))
    return pyramid

```

### How to produce a triangle

```python
def format_pyramid(ps: [str]) -> str:
    pyramid = []
    for p in ps:
        pyramid.append(p.center(WIDTH))
    return CRLF.join(pyramid)
```

```python
print(format_pyramid(make_pyramid(7)))
```

```python
def format_upside_down_pyramid(ps: [str]) -> str:
    pyramid = []
    for p in ps[::-1]:
        pyramid.append(p.center(WIDTH))
    return CRLF.join(pyramid)
```

```python
print(format_upside_down_pyramid(make_pyramid(7)))
```

```python
def format_rt_triangle(ps: [str]) -> str:
    pyramid = []
    for p in ps:
        pyramid.append(p)
    return CRLF.join(pyramid)
```

```python
print(format_rt_triangle(make_pyramid(7)))
```

```python
def format_mirror_rt_triangle(ps: [str]) -> str:
    pyramid = []
    for p in ps:
        pyramid.append(p.rjust(WIDTH))
    return CRLF.join(pyramid)
```

```python
print(format_mirror_rt_triangle(make_pyramid(7)))
```

```python
def format_diamond(ps: [str]) -> str:
    pyramid = []
    for p in ps + ps[:-1][::-1]:
        pyramid.append(p.center(WIDTH))
    return CRLF.join(pyramid)
```

```python
print(format_diamond(make_pyramid(7)))
```
