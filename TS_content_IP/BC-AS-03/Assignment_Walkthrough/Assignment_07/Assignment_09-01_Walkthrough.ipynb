Write a program that produces a tableau like this


$~13~~17~+$

$~~6~~34~\times$

$~~3~~68~+$

$~~1~136~+$

$=221$ 


Remember that the algorithm is : Halve and double; Add the $+$s, skip the $\times$s. The purpose is to also do a bit of output formatting

```python
def rus_mult_print(a: int, b: int) -> None:
    '''Remember we have to produce the tableau
    not just compute the product'''
    PLUS, CROSS, SPACE, EQUALS = '+', 'X', ' ', '='
    ab = 0
    while a > 0:
        if a % 2 == 0:
            sign = CROSS
        else:
            sign = PLUS
            ab += b
        print(f'{a: 4d} {b: 6d}' + SPACE + sign)
        a //= 2
        b *= 2
    print(EQUALS + f'{ab:10}')
    return
```

```python
rus_mult_print(13, 17)
```

```python
def rus_mult(a: int, b: int) -> [str]:
    '''Remember we have to produce the tableau
    not just compute the product'''
    PLUS, CROSS, SPACE, EQUALS = '+', 'X', ' ', '='
    ab = 0
    tableau = []
    while a > 0:
        if a % 2 == 0:
            sign = CROSS
        else:
            sign = PLUS
            ab += b
        tableau.append(f'{a: 4d} {b: 6d}' + SPACE + sign)
        a //= 2
        b *= 2
    tableau.append(EQUALS + f'{ab:10}')
    return tableau
```

```python
print('\n'.join(rus_mult(17, 12)))
```
