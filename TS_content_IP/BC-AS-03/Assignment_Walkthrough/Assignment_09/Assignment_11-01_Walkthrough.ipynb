## Assignment_11-01

Implement a function that converts roman number to arabic number

Example:

 > roman2arabic("V") --> 5

 > roman2arabic("IX") --> 9





```python
def roman2arabic(r: str) -> int:
    arabics = {'M':1000,'D':500,'C':100,'L':50,'X':10,'V':5,'I':1}
    a = 0
    r = r.upper().replace("IV", "IIII").replace("IX", "VIIII").replace("XL", "XXXX").replace("XC", "LXXXX").replace("CD", "CCCC").replace("CM", "DCCCC")
    for i in r:
        a += arabics[i]
    return a
```
