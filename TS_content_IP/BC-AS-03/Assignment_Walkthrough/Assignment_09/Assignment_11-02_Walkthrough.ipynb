## Assignment_11-02

program to convert arabic to roman

Example:

> arabic2roman(5) --> V

> arabic2roman(9) --> IX






```python
def arabic2roman(a:int) -> str:
    romans = {1000:'M',900:'CM',500:'D',400:'CD',100:'C',90:'XC',50:'L',40:'XL',10:'X',9:'IX',5:'V',4:'IV',1:'I'}
    r = ""
    for n in sorted(romans.keys(),reverse = True):
        while a >= n:
            r += romans[n]
            a -= n
    return r
```

One more example of how organizing the data makes for easy code
