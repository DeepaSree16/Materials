Write a function that Converts vowels into **upper** case character and consonants into **lower** case character in a given string .

```python
#Version -1
def vowel_to_upper_cons_to_lower(string: str)-> str:
    converted_string = ''
    for char in string:
        if char in 'aeiouAEIOU':
            converted_string += char.upper()
        else:
            converted_string += char.lower()
    return converted_string

vowel_to_upper_cons_to_lower("pYthon is Fun")
```

```python
#Version -2
def is_vowel(letter: str)-> bool:
  return letter in 'aeiouAEIOU'

def convertcase(text: str)-> str:
  return ''.join([char.upper() if is_vowel(char) else char.lower() for char in text])
```

```python
convertcase("python is fun")
```
