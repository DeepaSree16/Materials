Print integers 1 to N, but print “Fizz” if an integer is divisible by 3, “Buzz” if an integer is divisible by 5, and “FizzBuzz” if an integer is divisible by both 3 and 5.

Of course we do not want to sprinkle print statements, so let us generate them.

```python
#Version 1
def fizz_buzz(n: int)-> list([str]):
    fizz_buzz_seq = []
    for i in range(1, n + 1):
        if i % 3 == 0 and i % 5 == 0:
            fizz_buzz_seq.append("FizzBuzz")
        elif i % 3 == 0:
            fizz_buzz_seq.append("Fizz")
        elif i % 5 == 0:
            fizz_buzz_seq.append("Buzz")
        else:
            fizz_buzz_seq.append(str(i))
    return fizz_buzz_seq
print(fizz_buzz(50))
```

```python
def fb(n: int) -> str:
    if n % 15 == 0:
        return "FIZZBUZZ"
    elif n % 3 == 0:
        return "FIZZ"
    elif n % 5 == 0:
        return "BUZZ"
    else:
        return str(n)

def fizzbuzz(below):
    fbseq = []
    for i in range(1, below):
        fbseq.append(fb(i))
    return fbseq
fizzbuzz(16)
```

```python
#Version 3

all_fbs = ["FIZZBUZZ", '', '', "FIZZ", '', "BUZZ", "FIZZ", '', '', "FIZZ", "BUZZ", '', "FIZZ", '', '']

def fb2(n: int) -> str:
    position = n % 15
    if all_fbs[position] == '':
        return str(n)
    else:
        return all_fbs[position]

def fizz_buzz2(below: int)  -> [str]:
    fbseq = []
    for i in range(1, below):
        fbseq.append(fb2(i))
    return fbseq
```
