Write a function that classifies given list of numbers into **Even** and **Odd** Sublists.

```python
#Version 1

def split_Even_Odd_Sublist(num_array: [int]) -> (int, int):
   evenlist = [] 
   oddlist = [] 
   for i in num_array: 
      if i % 2 == 0: 
         evenlist.append(i) 
      else: 
         oddlist.append(i)
   return evenlist, oddlist
num_array = [2,4,6,8,1,3,5,7,9,11,12,14,15,21]

print(split_Even_Odd_Sublist(num_array))
```
