###Count vowels and consonants

Write a program to read the content from the "Content.txt" file and return the count of vowels and consonants.

* Content.txt 

https://drive.google.com/file/d/1h60E9b3r9ecFLPxnd5lYzDzLiza0NsK3/view?usp=sharing

```python
def read_file(filename: str) -> [str]:
    data = open(filename , "r")
    return data.readlines()

print(read_file("Content.txt"))
```

```python
def is_vowel(char: str) -> bool:
    return char in "aeiouAEIOU"
```

```python
def count_vowel_and_cons(filename: str)-> (int, int):
    data = read_file(filename)
    vowels = "AEIOUaeiou"
    vowel, cons = 0, 0
    for line in data:
        for char in line:
            if char.isalpha():
                if is_vowel(char):
                    vowel += 1
                else:
                    cons += 1
    return vowel, cons
    

count_vowel_and_cons("Content.txt")
```
