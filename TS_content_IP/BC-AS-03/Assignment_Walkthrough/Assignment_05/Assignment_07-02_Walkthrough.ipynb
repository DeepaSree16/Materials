Convert a variable name from camel case

```python
#Version 1
UNDERSCORE = '_'
def camel_case_to_snake_case(camel: str) -> str:
    snake_case = camel[0].lower()
    for char in camel[1:]:
        if char.isupper():
            snake_case += UNDERSCORE + char.lower()
        else:
            snake_case += char
    return snake_case

camel_case_to_snake_case("learnTechWiseTalentsprint")  
```

```python
def helper(char: str) -> str:
    if char.isupper():
        return "_" + char.lower()
    else:
        return char

def camel_to_snake(camel: str) -> str:
    snake = camel[0].lower()
    for ch in camel[1:]:
        snake += helper(ch)
    return snake
camel_to_snake("HelloWorldPython")
```

```python
# For reference
def camel_case_to_snake_case2(camel: str)-> str:
    return camel[0] + ''.join([helper(char) for char in camel[1:]])

camel_case_to_snake_case2("learnTechWiseTalentsprint")  
```
