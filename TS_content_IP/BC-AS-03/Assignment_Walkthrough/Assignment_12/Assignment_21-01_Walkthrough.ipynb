```python
#version-1
def fb(k: int)-> str:
    if k % 15 == 0:
        return "FIZZBUZZ"
    elif k % 3 == 0:
        return "FIZZ"
    elif k % 5 == 0:
        return "BIZZ"
    else:
        return str(k)

def fizz_buzz(k: int)-> [str]:
    return map(fb, range(1, k+1))

print(fizz_buzz(15))
print(list(fizz_buzz(15)))
```

```python
#version-2
def fb(k: int)-> str:
    return "FIZZBUZZ" if k % 15 == 0 else "FIZZ" if k % 3 == 0 else "BUZZ" if k % 5 == 0 else str(k)

def fizz_buzz(k: int)-> [str]:
    return map(fb, range(1, k+1))

print(list(fizz_buzz(15)))
```
