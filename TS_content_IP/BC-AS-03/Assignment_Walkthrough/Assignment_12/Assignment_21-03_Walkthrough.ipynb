```python
ARROW = "->"

def repeat_sequence(s: str)-> str:
    seq = []
    for item in list(s):
        seq.append(item.upper() + item.lower() * s.index(item))
    return ARROW.join(seq)

print(repeat_sequence('ab'))
print(repeat_sequence('UVWX'))
```

```python
ARROW = "->"

def repeat_sequence(s: str)-> str:
    seq = []
    for i, ch in enumerate(s, start = 1):
        seq.append((ch * i).capitalize())
    return  ARROW. join(seq)

print(repeat_sequence('ab'))
print(repeat_sequence('UVWX'))
```

```python
#with map
ARROW = "->"
def make_seq(char: str)-> str:
    return char[1].upper() + char[1].lower() * char[0]

def repeat_sequence(string: str) -> str:
    return ARROW.join(map(make_seq, enumerate(string)))

print(repeat_sequence("ab"))
print(repeat_sequence("UVWX"))
```
