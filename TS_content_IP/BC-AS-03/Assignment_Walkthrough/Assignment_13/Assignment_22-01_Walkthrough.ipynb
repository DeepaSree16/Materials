```python
#version - 1
def fib(n: int)-> [int]:
    a, b = 0, 1
    fibs = []
    while a < n:
        fibs.append(a)
        a, b = b, a+b
    return fibs

def euler2(limit: int)-> [int]:
    return sum(filter(lambda x: x % 2 == 0, fib(limit)))

euler2(4000000)
```

```python
#version - 2
def fib(n):
    a, b = 0, 1
    while a < n:
        yield a
        a, b = b, a + b

def euler2(limit: int)-> [int]:
    return sum(filter(lambda x: x % 2 == 0, fib(limit)))

euler2(4000000)
```

```python
#version - 3
def fib(n):
    a, b = 0, 1
    while a < n:
        yield a
        a, b = b, a + b

def euler2(limit: int)-> [int]:
    return sum([x for x in fib(limit) if x % 2 == 0])

euler2(4000000)
```
