```python
def split_even_odd_sublist(num_list: [int]) -> ([int], [int]):
    return [i for i in num_list if i % 2 == 0], [i for i in num_list if i % 2 != 0]
```

```python
split_even_odd_sublist([2, 4, 6, 8, 1, 3, 5, 7, 9, 11, 12, 14, 15, 21])
```

```python
def split_even_odd_sublist1(num_list: [int]) -> ([int], [int]):
    return list(filter(lambda x: x % 2 == 0, num_list)), list(filter(lambda x: x % 2 != 0, num_list))
```
