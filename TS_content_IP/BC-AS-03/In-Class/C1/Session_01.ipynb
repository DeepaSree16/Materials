## Problem 1
### Euler Problem 4

```python
def is_palindrome(n: int) -> bool:
    s = str(n)
    return s == s[::-1]
def euler004(start: int, limit: int) -> int:
    prods = []
    for a in range(start, limit):
        for b in range(a, limit):
            prod = a * b
            if is_palindrome(prod):
                prods.append(prod)
    return max(prods)
```

```python
euler004(10, 100)
```

```python
# Euler Problem 1

def euler001(limit: int) -> int:
    k, total = 1, 0
    for k in range(1, limit):
        if k % 3 == 0 or k % 5 == 0:
            total += k
    return total
# Compare with the while version
    
```

```python
LIMIT = 1000
euler001(LIMIT)
```

```python
# Euler Problem 1
def select(n: int) -> bool:
    return (n % 3) * (n % 5) == 0

def euler001a(limit: int) -> int:
    k, total = 1, 0
    for k in range(1, limit):
        if select(k):
            total += k
    return total
# Compare with the while version
```

```python
LIMIT = 1000
euler001a(LIMIT)
```

```python
# Euler Problem 1
def multiply_with(n: int) -> bool:
    if n % 3 == 0 or n % 5 == 0:
        return 1
    return 0

def euler001b(limit: int) -> int:
    k, total = 1, 0
    for k in range(1, limit):
        total += k * multiply_with(k)
    return total
# Is this good? If so why so? If not why not?
```

## Problem 3

Given a string of the form "mmddyyyy", (example: "02072016") compute the seven digit number yyyyddd, where ddd is the $n^{th}$ days of the year. For example  "02072016" is the $38^{th}$ day of the year and 2016038 is the expected result; similarly "12201996" -> 1996355



```python
DAYS_IN_MONTH = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
```

```python
def nthday(s: str) -> int:
    mm = int(s[:2])
    dd = int(s[2:4])
    yyyy = int(s[4:])
    ddd = 0
    for i in range(mm):
        ddd += DAYS_IN_MONTH[i]
    return yyyy * 1000 + ddd
```

```python
DAYS_IN_MONTH =   [0, 31, 28, 31, 30,  31,  30,  31,  31,  30,  31,  30,  31]
CUMULATIVE_DAYS = [0,  0, 31, 59, 90, 120, 151, 181, 212, 243, 273, 304,  334]
```

```python
def nth_day(s: str) -> int:
    mm = int(s[:2])
    dd = int(s[2:4])
    yyyy = int(s[4:])
    return yyyy * 1000 + dd + CUMULATIVE_DAYS[mm]
```

The idea is to do with data what can be done in code

But in both cases we are not accounting for leap years.

```python
CUMULATIVE_DAYS = [[0,  0, 31, 59, 90, 120, 151, 181, 212, 243, 273, 304,  334],
                   [0,  0, 31, 60, 91, 121, 152, 182, 213, 244, 274, 305,  335]]               
def leap_year_adj(y: int) -> int:
    if y % 100 == 0:
        yy = y // 100
    else:
        yy = y % 100
    if yy % 4 == 0:
        return 1
    else:
        return 0

def mmddyyyy_to_yyyyddd(s: str) -> int:
    mm = int(s[:2])
    dd = int(s[2:4])
    yyyy = int(s[4:])  
    return yyyy * 1000 + dd + CUMULATIVE_DAYS[leap_year_adj(yyyy)][mm]
```

Once again we are using the "DATA IS CODE" principle :-)
