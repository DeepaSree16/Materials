##Counting vowels

Implement a function that takes a string as an argument and returns the count of vowels contained within it.

Note: 
- a, e, i, o, u are considered as vowels

**Examples**

vowel_count("Celebration") ➞ 5

vowel_count("Prediction") ➞ 4

```python
def isvowel(char: str) -> bool:
    return char in "aeiouAEIOU"

def vowel_count(txt:str) -> int:
    countofvowels = 0
    for letter in txt:
        if isvowel(letter):
          countofvowels += 1
    return countofvowels

print(vowel_count("TechWise"))
```

## Counting characters

* Implement a function to count vowels, consonants, digits and special characters in a given string

```python
def is_vowel(char: str) -> bool:
    return char in "aeiouAEIOU"

def count_characters(text: str)->(int):
    vowels = 0
    consonants = 0
    digits = 0
    specials = 0

    for letter in text:
        if letter.isalpha():
            if is_vowel(letter):
                vowels += 1
            else:
                consonants += 1
        elif letter.isdigit():
            digits += 1
        elif not letter.isspace():
            specials  += 1
    return vowels, consonants, digits, specials
    


text = '''learn.techwise.2022@talentsprint.com'''
count_characters(text)
```

#ROT13

Implement a function which accepts a string as an argument and generates the ROT13 Caesar Cipher

The ROT-13 (i.e. rotate by 13 places) method is a very simple encryption method, in which one replaces each letter of the alphabet with the 13th letter after it. So, ‘A’ becomes ‘N’, ‘B’ becomes ‘M’ and so on. This method is a special case of the Caesar Cipher, with which Caesar encoded his letters in the ancient Rome.

Here is our approach for encoding a given string:

* check if given input has all valid characters, i.e every character is an  alphabet [a - z or A - Z] - fucntion is_string() returns "True" if the char is an alphabet.

* as the next step we need to add 13 to  character and encode the character- rot13() returns the encoded character of the input character

* and now we generate the encoded string using the function - generate_encrypted_string() which returns encoded string

*need to discuss about the ascii values and also functions like ord() and chr()



But this is very poor and non-pythonic code. We should use built-in string methods and not rely on ascii values and such.

```python
def rot13(text):
   locase = "abcdefghijklmnopqrstuvwxyzabcdefghijklm"
   upcase = locase.upper()
   letters = locase + upcase
   #print(letters)
   encoded_text = ""

   for char in text:
       encoded_text += letters[(letters.find(char) + 13)]
   return encoded_text


text = "Tech Wise"
print(rot13(text))
text1 = "TechWise"
print(rot13(text1))
```

## Check if a Sentence is Palindrome 

Implement a fumction to check if a sentence is a palindrome or not. 

Note: ignore white spaces and other characters to consider sentences as palindrome.

Example:

is_palindrome("A man, a plan, a canal: Panama") ➞ True

Approch:

* we need to check if the given sentence is a palindrome 
* points to be noted: 
  - convert all the characters in the sentence to the same case, here we are converting the sentence to lowercase
  - ignore spaces and other special characters
  - consider only alphabets

* discuss about string methods
* discuss about string striding
* reversing a string is to slice between default limits (by omitting the first and last indexes) with a stride of -1



```python
def is_palindrome(text: str) -> bool:
    text = text.lower()
    sentence = ''
    for char in text:
        if 'a' <= char <= 'z':
            sentence += char
    return sentence == sentence[::-1]

print(is_palindrome("A man,  a plan, a canal: Panama"))
```

```python
def is_palindrome(s: str) -> bool:
  new_string = ''
  for char in s.lower():
    if char.isalpha():
      new_string += char
  return new_string == new_string[::-1]

print(is_palindrome("A man,  a plan, a canal: Panama"))
```
