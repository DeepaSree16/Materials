### What are the observations?

* All are *is*_something
* All are checking a property of numbers, that is strictly increasing:
    - that is if $a > b$ then for all these functions $f(a) > f(b)$
* Duplication in **structure** of the code
* How do we fix this? How do we take advantage of this?
    - Move the structure of the duplication: namely loop, apply and verify, increment to a function
    - Call this as function with the required  function as argument
 

```python
# Here is the structure in a function
def is_f(f, n: int) -> bool:
        r = 0
        while f(r) < n:
            r += 1
        return f(r) == n
```

```python
def triangular(r: int):
        return (r * (r + 1)) // 2

def istriangular(n: int) -> bool:
        return is_f(triangular, n)
```

```python
def square(k: int) -> int:
    return k * k

def is_square(p: int) -> bool:
    return is_f(square, p)
```

```python
istriangular(15)
```

```python
istriangular(12)
```

# How to add certain value to each item in any iterator

* Say marks of students are available in an iterator (say list) and want to add 3 as grace marks to all

```python
# normal way

marks = [70 , 60, 65, 80, 55]
updated_marks = [ ]
for mark in marks:
    updated_marks.append(mark + 3)
updated_marks


```

**Use Higher Order Function: map**

* map(f, l) -> f(l[0]), f(l[1]), .... f(l[-1])

* Applying the same rule for all elements

* Returns same size as input size

* Return data type is same as return type of function

* Map returns map object, need to type cast to say list to view


```python
def grace_marks(x: int) -> int:
        return x + 3

updated_marks = map(grace_marks, marks)

print(updated_marks)
print(list(updated_marks))
```

**How to remove certain items?**

How to remove items from an iterator

which does not satisfy given condition?

Say we have list of marks in a list and want to have only more than pass marks,say 40

```python
#Syntax: Usual way

marks = [50 , 60, 35, 39, 70, 75, 85, 30]
update_marks = [ ]
for mark in marks:
    if mark > 40:
        updated_marks.append(mark)

updated_marks
```

**Use Higher Order Function: filter**

filter(f, l) -> f(l[x]), say, item x, which satisfies given condition

Returns subset or part of provided iterator or container

Filter returns filter object, need to type cast to required data type (say list) to
view


```python
def passed(x: int) -> bool:
        return x > 40

marks = [50, 60, 35, 39, 70, 75, 85, 30]
updated_marks = filter(passed, marks)

print (updated_marks)
print(list(updated_marks))
```

# Problem 01:

Put the digits of a number into a list

```python
def num_to_digits(n: int) -> [int]:
    digits = []
    while n > 0:
        digits.insert(0, n % 10)
        n //= 10
    return digits

def num2digits(n: int) -> [int]:
    return list(map(int, str(n)))
```

```python
print(num_to_digits(1729))
print(num2digits(1729))
```

# Problem 03

Count vowels, consonants and special characters


```python
def is_vowel(ch: str) -> bool:
    return ch in "aeiouAEIOU"

def is_consonant(ch: str) -> bool:
    return ch.isalpha() and (not is_vowel(ch))

def is_special(ch: str) -> bool:
    return (not ch.isalnum()) and (not ch.isspace())
```

```python
def howmany(s: str) -> (int, int):
    return len(list(filter(is_vowel, s))), len(list(filter(is_consonant, s))), len(list(filter(is_special, s)))
```
