###Kaprekar sequence

It is the sequence where the next kaprekar generated by the subtraction of biggest and smallest numbers formed by the arrangements of the given number.
The sequence will stop if the same number appear twice.

For Example:

Sample Input: 9541

Sample Output: [9541, 8082, 8532, 6174]

Explanation:

9541 - 1459 = 8082

8820 - 0288 = 8532

8532 - 2358 = 6174

7641 - 1467 = 6174

```python
def num2digit(n: int)->[int]:
    return [int(ch) for ch in str(n)]

num2digit(9541)
```

```python
def digits2num(digits: [int])->int:
    num = 0
    for digit in digits:
        num = num * 10 + digit
    return num
digits2num([9, 5, 4, 1])
```

```python
def biggest(n: int)->int:
    return digits2num(sorted(num2digit(n), reverse = True))
```

```python
def smallest(n: int)->int:
    return digits2num(sorted(num2digit(n)))
```

```python
def next_kaprekar(n: int)->int:
    return biggest(n) - smallest(n)
```

```python
def kaprekar_seq(n: int)->[int]:
    k = next_kaprekar(n)
    if n == k:
        return [n]
    else:
        return [n] + kaprekar_seq(k)
```

```python
kaprekar_seq(9541)
```

```python
kaprekar_seq(3954)
```
