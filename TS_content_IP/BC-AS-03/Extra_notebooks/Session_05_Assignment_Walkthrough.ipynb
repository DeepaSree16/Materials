```python
#version1

numerals = {
    'I': 1,
    'V': 5,
    'X': 10,
    'L': 50,
    'C': 100,
    'D': 500,
    'M': 1000
}
 
def Roman_to_Arabic(roman):
    total = 0
 
    for i in range(len(roman) - 1):
        left = roman[i]
        right = roman[i + 1]
        if numerals[left] < numerals[right]:
            total -= numerals[left]
        else:
            total += numerals[left]
    total += numerals[roman[-1]]
 
    return total
 
print(Roman_to_Arabic('XXVII'))
```

```python
#Version 2: using upper() and replace methods
def roman2arabic(r:str) -> int:
    arabics = {'M':1000,'D':500,'C':100,'L':50,'X':10,'V':5,'I':1}
    arabic_number = 0
    r = r.upper().replace("IV", "IIII")
    r = r.upper().replace("IX", "VIIII")
    r = r.upper().replace("XL", "XXXX")
    r = r.upper().replace("XC", "LXXXX")
    r = r.upper().replace("CD", "CCCC")
    r = r.upper().replace("CM", "DCCCC")
    for i in r:
        arabic_number += arabics[i]
    return arabic_number
roman2arabic('XXVIII')
```

```python
#version 3: using list comprehension
romans = {'M': 1000, 'D': 500, 'C': 100, 'L': 50, 'X': 10, 'V': 5, 'I': 1}
```

```python
def Roman2Arabic(roman: str)->int:
  std_roman = roman.replace('CM', 'DCCCC'). replace('CD','CCCC').replace('XC', 'LXXXX').replace('XL', 'XXXX').replace('IX', 'VIIII'). replace('IV', 'IIII')
  return sum([romans[rc] for rc in std_roman])
```

```python
#Version 2
def arabic2roman(a:int) -> str:
    romans = {1000:'M',900:'CM',500:'D',400:'CD',100:'C',90:'XC',50:'L',40:'XL',10:'X',9:'IX',5:'V',4:'IV',1:'I'}
    roman_number= ""
    for ch in sorted(romans.keys(),reverse = True):
        while a >= ch:
            roman_number += romans[ch]
            a -= ch
    return roman_number

print(arabic2roman(800))
```
