##Required Arguments

###Implement a function which calculate the sum of digits of a number.

```python
def sum_of_digits(num):
 total = 0
 while num > 0:
   total += num % 10
   num //= 10
 return total

sum_of_digits(1234)
```

## Default Arguments

Implement a function to calculate the gross salary of an employee for the month of March. He is expecting 10% of HRA and 12% of TA on his basic salary. This month every employee will receive a $500 as a bonus. 

```python
def calculate_gross(basic_salary: int, bonus = 500)-> float:
  ta = basic_salary * 0.1
  da = basic_salary * 0.12
  return basic_salary + ta + da + bonus

calculate_gross(3000)
```

##Variable-length Arguments (*args)

Calculate the sum of given numbers.

```python
def calculate_sum(*nums):
  total = 0
  #print(type(nums))
  for item in nums:
    total += item
  return total

print(calculate_sum(12, 10, 14, 56, 90))
print(calculate_sum(10, 20, 30))
print(calculate_sum(1, 2, 3, 4, 5, 6, 7, 8, 9))
```

#Higher Order Functions

##Euler Problem - 1

If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Find the sum of all the multiples of 3 or 5 below 1000.

```python
#Method-1
def Euler_1(limit):
  return sum(filter(lambda x: x % 3 == 0 or x % 5 == 0, range(LIMIT)))

Euler_1(1000)
```

```python
#Method-2
def select(x):
    return x % 3 == 0 or x % 5 == 0

def Euler_01(limit):
  return sum(filter(select, range(LIMIT)))

Euler_01(1000)
```

#List Comprehension

###Problem 1

Implement a function to repeat the sequence as per the given order.

**Example**
 - repeatSequence('ab') => A->Bb
 - repeatSequence('UVWX') => U->Vv->Www->Xxxx

```python
def repeatSequence(s: str)-> str:
  return "->".join([item.upper() + item.lower() * s.index(item) for item in list(s)])

print(repeatSequence('ab'))
print(repeatSequence('UVWX'))
```

```python
def repeatSequence(s: str)-> str:
  return  "->". join([(ch * i).capitalize() for i, ch in enumerate(s, start = 1)])

print(repeatSequence('ab'))
print(repeatSequence('UVWX'))
```

##Problem - 3

Generate the Amstrong numbers between two given numbers.

```python
def is_armstrong(n: int)->bool:
  return sum([int(i) ** len(str(n)) for i in str(n)]) == n
```

```python
def makeArmstrong(start: int, limit: int):
  return filter(is_armstrong, range(start, limit))
```

```python
list(makeArmstrong(100, 1000))
```
