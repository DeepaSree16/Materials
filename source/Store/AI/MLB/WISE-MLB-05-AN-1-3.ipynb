**Pre-processing**

```python
#start replaceTwoOrMore
def replaceTwoOrMore(s):
    #look for 2 or more repetitions of character and replace with the character itself
    pattern = re.compile(r"(.)\1{1,}", re.DOTALL)
    return pattern.sub(r"\1\1", s)
```

```python
#start process_review
def processReview(review):
    #Convert to lower case
    review = review.lower()
    review = review.translate(str.maketrans('', '', string.punctuation))
    # Removing numbers
    review = re.sub('[0-9]', '', review) 
    #Remove additional white spaces
    review = re.sub('[\s]+', ' ', review)
    #Replace #word with word
    review = re.sub(r'#([^\s]+)', r'\1', review)
    #trim
    review = review.strip('\'"')
    review = review.strip('.,')
    review = replaceTwoOrMore(review)
    return review
```

```python
processedReviews = []
for review in reviews:
  processedReviews.append(processReview(review))
```

```python
processedReviews
```

```python
def getNGrams(wordlist, n):
    ngrams = []
    for i in range(len(wordlist)-(n-1)):
        ngrams.append(wordlist[i:i+n])
    return ngrams
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(ngram_range=(2, 2))
X = count_vect.fit_transform(processedReviews)
```

```python
print(count_vect.get_feature_names())
```

```python
print(pd.DataFrame(X.A[:, :10], columns=count_vect.get_feature_names()[:10]).to_string())
```

```python
import numpy as np
from sklearn.decomposition import PCA
model = PCA(n_components=1000)
reduced_X = model.fit_transform(X.A)
```

```python
print(np.cumsum(model.explained_variance_ratio_)*100)
```
