```
# Stories_list    UniqueWords    AvgSentLength     Label 
#     1               x1               x2            y 

####################### 1. GET PUNCTUATION COUNT #######################
import string
from nltk.collocations import BigramCollocationFinder
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
stop_words = set(stopwords.words('english'))

def getPunctuationCount(story):
  count = 0
  for i in story:
      if i in string.punctuation:
          count+=1
  return count


####################### 2.GET UNIQUE WORDS COUNT #######################
def getStripped(story):
  table = str.maketrans('', '', string.punctuation)
  words = story.replace("\r\n"," ").replace("\n"," ").strip().split(" ")
  stripped = [w.translate(table) for w in words]
  while ("" in stripped):
    stripped.remove("")
  return stripped

def getFilteredWords(word_tokens):
  filtered_sentence = ['number' if w.isdigit() else w.lower() for w in word_tokens if not w in stop_words and len(w) > 2]
  return filtered_sentence

def getUniques(story):
  return len(set(getFilteredWords(getStripped(story))))


####################### 3. GET AVERAGE SENTENCE LENGTH #######################
def getAvgSentenceLen(story):
  while (".." in story):
    story = story.replace("..",'.')    
  dummy = story.split(".")
  quantity = len(dummy)
  total_sum = 0
  for i in dummy:
    total_sum += len(i)
  return total_sum//quantity

####################### 4. GET BI-GRAMS COUNT #######################
def biGrams(story):
  bigram_measures = nltk.collocations.BigramAssocMeasures()
  finder = BigramCollocationFinder.from_words(getStripped(story))
  scored = finder.score_ngrams(bigram_measures.raw_freq)
  return sorted(bigram for bigram, score in scored)


####################### 5. GET ADJECTIVE COUNT #######################
def getAdjCount(story):
  count = 0
  for i in nltk.pos_tag(getStripped(story)):
    if i[1]=="JJ":
      count+=1
  return count

####################### 6. GET FREQUENCY DISTRIBUTION #######################
def freqDistribution(story):
  freqDist = nltk.FreqDist(story)
  return sum(freqDist.values())//len(freqDist)
```

```
punctuations_1 = pd.Series([getPunctuationCount(i) for i in FirstBook], name='punctuations')
punctuations_2 = pd.Series([getPunctuationCount(i) for i in SecondBook], name='punctuations')

uniques_1 = pd.Series([getUniques(i) for i in FirstBook], name='uniques')
uniques_2 = pd.Series([getUniques(i) for i in SecondBook], name='uniques')

avgSentenceLen_1 = pd.Series([getAvgSentenceLen(i) for i in FirstBook], name='avg_Sentence_Len')
avgSentenceLen_2 = pd.Series([getAvgSentenceLen(i) for i in SecondBook], name='avg_Sentence_Len')

biGrams_1 = pd.Series([len(biGrams(i)) for i in FirstBook], name='biGrams')
biGrams_2 = pd.Series([len(biGrams(i)) for i in SecondBook], name='biGrams')

AdjCount_1 = pd.Series([getAdjCount(i) for i in FirstBook], name='Adjectives')
AdjCount_2 = pd.Series([getAdjCount(i) for i in SecondBook], name='Adjectives')

Frequency_1 = pd.Series([freqDistribution(i) for i in FirstBook], name='Frequency')
Frequency_2 = pd.Series([freqDistribution(i) for i in SecondBook], name='Frequency')

AuthorName_1 = pd.Series([getAuthor(FirstAuthor) for i in FirstBook], name='AuthorName')
AuthorName_2 = pd.Series([getAuthor(SecondAuthor) for i in SecondBook], name='AuthorName')
```

```
df_1 = pd.concat([punctuations_1, uniques_1, avgSentenceLen_1, biGrams_1, AdjCount_1, Frequency_1, AuthorName_1], axis=1)
df_2 = pd.concat([punctuations_2, uniques_2, avgSentenceLen_2, biGrams_2, AdjCount_2, Frequency_2, AuthorName_2], axis=1)

final_Df = pd.concat([df_1,df_2],ignore_index=True)
final_Df.head()
```

**CountVectorizer**

```
# YOUR CODE HERE (HINT: Convert to numpy array if needed)

from sklearn.feature_extraction.text import CountVectorizer

def countVec(story):
  cleanedStory = [' '.join(getFilteredWords(getStripped(sentence))) for sentence in story]
  vectorizer = CountVectorizer()
  story_vectors = vectorizer.fit_transform(cleanedStory)  
  return story_vectors.toarray()

countVectorizer = pd.DataFrame(countVec(FirstBook + SecondBook))
print("Shape of the Countvec Dataframe: ", countVectorizer.shape)
countVectorizer.head()
```

**TfidfVectorizer**

```
from sklearn.feature_extraction.text import TfidfVectorizer

def tfIdf(story):
  cleanedStory = [' '.join(getFilteredWords(getStripped(sentence))) for sentence in story]
  vectorizer = TfidfVectorizer(norm = False, smooth_idf = False)
  story_vectors = vectorizer.fit_transform(cleanedStory)
  return story_vectors.toarray()

tfIdf_Vectorizer = pd.DataFrame(tfIdf(FirstBook + SecondBook))
tfIdf_Vectorizer.shape
print("Shape of the TFIDF Vec Dataframe: ", tfIdf_Vectorizer.shape)
tfIdf_Vectorizer.head()
```

**Word2Vec**

```
import gensim.models.keyedvectors as word2vec
model = word2vec.KeyedVectors.load_word2vec_format("/content/AIML_DS_GOOGLENEWS-VECTORS-NEGATIVE-300_STD.bin", binary=True, limit=500000)
```

```
def w2Vec(story, label):
  w2Vec_Features = []
  cleanedStory = [getFilteredWords(getStripped(sentence)) for sentence in story]
  allWords = [y for x in cleanedStory for y in x]
  for w in allWords:
    try:
      vectors = list(model[w])
      vectors.append(label)
      w2Vec_Features.append(vectors)
    except:
      pass
  return w2Vec_Features
```

```
vectors = [w2Vec(book, label) for label, book in enumerate([FirstBook,SecondBook])]
w2Vec_df = pd.DataFrame([y for x in vectors for y in x])

print("Shape of the WORD2Vec Dataframe: ", w2Vec_df.shape)
w2Vec_df.head()
```

```
w2Vec_Features = w2Vec_df.iloc[:, :-1]
w2Vec_Targets = w2Vec_df.iloc[:,-1]

print("Shape of Word2Vec Features: ", w2Vec_Features.shape)
print("Shape of Word2Vec Targets: ", w2Vec_Targets.shape)
```
