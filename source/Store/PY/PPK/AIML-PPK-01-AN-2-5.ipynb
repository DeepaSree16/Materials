### o) Remove "%" sign from the "Last % Hike" column values of the dataset

you can refer the below link :

https://stackoverflow.com/questions/52850192/python-extract-last-letter-of-a-string-from-a-pandas-column

```
df = df.rename(columns = {'Last % Hike':'Last Hike'})  ### NOT NECESSARY BUT, A VERY ROUTINE QUESTION FROM PARTICIPANTS AT THIS SEGMENT
df['Last Hike'] = [int(each[:-1]) for each in df['Last Hike']]
```
